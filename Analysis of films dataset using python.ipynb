### Replacing/Filling the missing values

```python
# Let's replace the missing values with zero
df1 = df.fillna(value=0)
df1
```

```python
# Let's replace integer missing values with 0 and string missing value to Xyz
df1 = df.fillna({'Length':0,
                'Popularity':0,
                'Subject': 'Xyz',
                'Actor':'Xyz',
                'Actress':'Xyz',
                'Director':'Xyz'})
df1
```

### Exploratory data analysis

##### 1)Find the movie with most awards.

```python
# Getting year from
df1['duration'] = df1['Year'].apply(lambda x: ('%Y'))
```

```python
# grouping year by Awards
df_temp = df.groupby('Year').sum()['Awards'].reset_index()
```

```python
# Setting the figure size
plt.figure(figsize=(80, 34))
plt.plot(df_temp['Year'], df_temp['Awards'], color='#b80045')
plt.xticks(rotation='vertical', size=10)
plt.show()
```

```python
# Grouping movies by awards
movie_awards = pd.DataFrame(df1.groupby('Title').sum()['Awards'])

# Sorting the dataframe in descending order
movie_awards.sort_values(by=['Awards'], inplace=True, ascending=False)

# Top 10 movies by awards
movie_awards[:10]
```

#### 2) Analyse the most popular movie.

```python
# Grouping movies by Popularity
popular_movie = pd.DataFrame(df1.groupby('Popularity'))
```

```python
# Sorting the dataframe in descending order
popular_movie.sort_values(by=['Popularity'], inplace=True, ascending=False)
```

```python
# Most popular movie
popular_movie[:10]
```

```python
# Grouping movies popularity
most_lengthy = pd.DataFrame(df1.groupby('Title')['Popularity'].count())
# Most popular
most_lengthy[:10]
```

```python
# Grouping movies length
most_lengthy = pd.DataFrame(df1.groupby('Title')['Length'].count())
# Movie with longest time duration
most_lengthy[:10]
```

#### 3) Analyse the relation between number of awards in a year and the most popular movie in a year


```python
# Grouping movies awards in relation to their popularity
most_lengthy = pd.DataFrame(df1.groupby('Title')['Length','Popularity'].count())
most_lengthy[:10]
```
